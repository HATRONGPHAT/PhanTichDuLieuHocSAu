```python
#1. Tạo Dataframe như dữ liệu đã cho trên.
import pandas as pd
import numpy as np

data = {
    "Name": ["Mr. Harris", "Mrs. Bradley", "Miss. Laina", "Mrs. Heath", "Mr. William", "Mr. James", "Mr. Timothy", "Mrs. Oscar", "Mrs. Nicholas", "Miss. Marguerite", "Kelly", "Mr. William", "Miss. Hulda", "Mr. Lawrence"],
    "Height": [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    "Weight": [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    "AGE": [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)
```

```python
#11.Kiểm tra có giá trị NaN nào trong DataFrame không
print("Các giá trị NaN trong DataFrame:")
print(df.isna().sum())
```

```python
#12 Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên.
df = df.drop_duplicates(subset=["Name"], keep="first")
print(df)
```

```python
#13 . Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó
mean_age = df[df["AGE"] <= 40]["AGE"].mean()
df["AGE"] = df["AGE"].apply(lambda x: mean_age if x > 40 else x)
print(df)
```

```python
#14.
df["cs_BMI"] = df["Weight"] / (df["Height"] ** 2)
print(df)
```

```python
#15 
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0
    elif 18.5 <= bmi < 23:
        return 1
    elif 23 <= bmi < 25:
        return 2
    else:
        return 3

df["Phan_Loai"] = df["cs_BMI"].apply(classify_bmi)
print(df)
```

```python
#16 Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1.
df["cs_BMI"] = (df["cs_BMI"] - df["cs_BMI"].min()) / (df["cs_BMI"].max() - df["cs_BMI"].min())
print(df)
```

```python
#17 
df["GT"] = df["Name"].apply(lambda x: "male" if "Mr" in x else "female")
df["Name"] = df["Name"].str.replace("Mr. |Mrs. |Miss. ", "", regex=True)  # Xóa tiền tố
print(df)
```

```python
#18 Vẽ biểu đồ tròn cho <GT>.
df["GT"].value_counts().plot.pie(autopct="%1.1f%%", colors=["blue", "pink"])
plt.title("Biểu đồ tròn cho GT")
plt.ylabel("")  # Ẩn nhãn cột y
plt.show()
```

```python
#19 Trực quan hóa dữ liệu người nữ trên <Phan_Loai>.
df_female = df[df["GT"] == "female"]
plt.bar(df_female["Name"], df_female["Phan_Loai"], color="purple")
plt.xticks(rotation=90)
plt.xlabel("Tên")
plt.ylabel("Phân loại BMI")
plt.title("Trực quan hóa Phân loại BMI của người nữ")
plt.show()
```
